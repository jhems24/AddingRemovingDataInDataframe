```python
#fill na values with zeros in df
df=df.fillna(0)
```

```python
#create a new column that displays amount per order
df['Amount']=df['PricePerItem']*df['Quantity']
```

```python
#filter DataFrame for list of customers who spend $50 per order
df_50=df[df['Amount']>=50]
```
